```python
import pandas as pd
df = pd.read_csv(r'/Users/suhaas/Machine-Learning/manu-sus-engine/Merged_dataset.csv')
df.head()
```

```python
df["Energy_Consumption_kWh"] = df["Avg_Power_Consumption"] * (df["Duration_Minutes"] / 60)
df.head()
```

```python
#0.82 kg CO2 per kWh  (India average)
df["Carbon_Emissions_kgCO2"] = df["Energy_Consumption_kWh"] * 0.82
df.head()
```

```python
targets = [
    "Hardness",
    "Friability",
    "Disintegration_Time",
    "Dissolution_Rate",
    "Content_Uniformity",
    "Energy_Consumption_kWh",
    "Carbon_Emissions_kgCO2"
]

X = df.drop(columns=targets + ["Batch_ID"])
y = df[targets]
```

```python
X.info()
```

```python
numeric_cols = X.select_dtypes(include=["int64","float64"]).columns
categorical_cols = X.select_dtypes(include=["object","category"]).columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

preprocessor = ColumnTransformer([
    ("num", StandardScaler(), numeric_cols),
    ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols)
])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
```

```python
rf_pipeline = Pipeline([
    ("preprocess", preprocessor),
    ("model", MultiOutputRegressor(
        RandomForestRegressor(
            n_estimators=400,
            max_depth=None,
            random_state=42,
            n_jobs=-1
        )
    ))
])
```

```python
%pip install xgboost
from xgboost import XGBRegressor
```

```python
xgb_pipeline = Pipeline([
    ("preprocess", preprocessor),
    ("model", MultiOutputRegressor(
        XGBRegressor(
            n_estimators=400,
            max_depth=6,
            learning_rate=0.1,
            random_state=42,
            n_jobs=-1
        )
    ))
])
```

```python
from sklearn.model_selection import cross_val_score, KFold

kf = KFold(n_splits=5, shuffle=True, random_state=42)

rf_scores = cross_val_score(rf_pipeline, X, y, cv=kf, scoring="r2")
xgb_scores = cross_val_score(xgb_pipeline, X, y, cv=kf, scoring="r2")
```
